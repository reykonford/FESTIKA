```python
semester = [
    "Rata-rata Semester 1",
    "Rata-rata Semester 2",
    "Rata-rata Semester 3",
    "Rata-rata Semester 4",
    "Rata-rata Semester 5"
]

def hitung_rata(nilai):
    return sum(nilai) / len(nilai)

def main():
    print("=== SISTEM PENGOLAHAN NILAI RAPORT KELAS XII ===")
    total_murid = int(input("Masukkan jumlah seluruh siswa kelas XII: "))

    nilairata = []

    for i in range(total_murid):
        print(f"\nNama siswa ke-{i+1}: ", end="")
        nama = input()

        nilai = []
        print("Masukkan nilai rata-rata per semester:")
        for j in range(5):
            while True:
                try:
                    n = float(input(f"{semester[j]}: "))
                    break
                except ValueError:
                    print("Input tidak valid. Harap masukkan angka.")
            nilai.append(n)

        rata = hitung_rata(nilai)

        nilairata.append({
            "nama": nama,
            "nilai": nilai,
            "rata": rata,
            "status": ""
        })

    nilairata.sort(key=lambda x: x["rata"], reverse=True)

    batas = (total_murid * 40) // 100

    for i in range(total_murid):
        if i < batas:
            nilairata[i]["status"] = "ELIGIBLE"
        else:
            nilairata[i]["status"] = "TIDAK ELIGIBLE"

    print("\n=== HASIL PERHITUNGAN DAN PERANKINGAN ===")
    print("RANK\tNAMA\t\tRATA-RATA\tSTATUS")

    for i, s in enumerate(nilairata):
        print(f"{i+1}\t{s['nama']}\t\t{s['rata']:.2f}\t\t{s['status']}")

main()
```
